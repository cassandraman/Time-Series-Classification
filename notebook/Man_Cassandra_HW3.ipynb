### C. Feature Extraction

#### i. Types of Time-Domain Features

Time-domain features are extracted from analysis of data with respect to time. Different time-domain features used in time-series classification can be grouped into the below categories:

Statistical Features: Time-domain features include values in the five-number summary (maximum, minimum, lower quartile, upper quartile, and median values), as well as the mean, standard deviation, skewness, and Kurtosis.

Time Series Features: Time-domain features include cross-correlations between each dimension, auto-correlations, parameters of the autoregressive model, and parameters of the moving average model.

Date Time Features: Time-domain features related to Date Time are created from observation time stamps, and include features such as minute, hour, season, and quarter.

Lag Features: Lag features describe values at timestamps that occured in the past. Window features are based on lag features, and provide summary statistics over a fixed period of previous timestamp data. Rolling window features provide a description over a range that includes the given data sample with other samples from before or after. Expanding window features includes features over all previous data.

Frequency Domain Features: Time-domain features include frequencies of amplitude peaks and quantiles for dimensions.

#### ii. Extract Time-Domain Features

```python
# Create dataframe for new dataset

feature_df_columns = ["Instance", "min1", "max1", "mean1", "median1", "SD1", "first_quartile1", "third_quartile1",
                      "min2", "max2", "mean2", "median2", "SD2", "first_quartile2", "third_quartile2",
                      "min3", "max3", "mean3", "median3", "SD3", "first_quartile3", "third_quartile3",
                      "min4", "max4", "mean4", "median4", "SD4", "first_quartile4", "third_quartile4",
                      "min5", "max5", "mean5", "median5", "SD5", "first_quartile5", "third_quartile5",
                      "min6", "max6", "mean6", "median6", "SD6", "first_quartile6", "third_quartile6"]
```

```python
all_files_list = []

# Set instance number 
counter = 1

# Iterate through all files
for filename in file_list:
    
    # Create list 
    instance_list = [counter]
    
    # Increment instance number
    counter = counter + 1

    # Read CSV
    curr_df = pd.read_csv(filename, skiprows=5, sep= '\s+|,', index_col = False, header=None, engine="python")
    curr_df.drop(0, axis=1, inplace=True)
    
    # Extract features from each time series
    for (columnName, columnData) in curr_df.iteritems():
        curr_list = columnData.values
        
        min_val = min(curr_list)
        max_val = max(curr_list)
        mean_val = statistics.mean(curr_list)
        SD_val = statistics.stdev(curr_list)
        quartile1_val = np.quantile(curr_list, 0.25)
        median_val = np.quantile(curr_list, 0.5)
        quartile3_val = np.quantile(curr_list, 0.75)
        timeseries_list = [min_val, max_val, mean_val, median_val, SD_val, quartile1_val, quartile3_val]
        instance_list.extend(timeseries_list)
    
    all_files_list.append(instance_list)
```

```python
# Create new dataset
feature_df = pd.DataFrame(all_files_list, columns=feature_df_columns)
pd.set_option('display.max_rows', None)
feature_df
```

#### iii. Estimate Standard Deviation and Build Bootstrap Confidence Interval

```python
# Estimate Standard Deviation

# Loop through each feature column
df_noinstance = feature_df.drop("Instance", axis=1)

SD_list = []

for column in df_noinstance:
    curr_feature = df_noinstance[column].tolist()
    
    # Calculate SD
    curr_feature_SD = statistics.stdev(curr_feature)
    SD_list.append(curr_feature_SD)

SD_df_columns = ["min1", "max1", "mean1", "median1", "SD1", "first_quartile1", "third_quartile1",
                 "min2", "max2", "mean2", "median2", "SD2", "first_quartile2", "third_quartile2",
                 "min3", "max3", "mean3", "median3", "SD3", "first_quartile3", "third_quartile3",
                 "min4", "max4", "mean4", "median4", "SD4", "first_quartile4", "third_quartile4",
                 "min5", "max5", "mean5", "median5", "SD5", "first_quartile5", "third_quartile5",
                 "min6", "max6", "mean6", "median6", "SD6", "first_quartile6", "third_quartile6"]

# Create Dataframe
SD_df = pd.DataFrame({"Feature": SD_df_columns, "Standard Deviation": SD_list})
SD_df
```

```python
# Build Bootstrap Confidence Interval

all_lower_list = []
all_upper_list = []
    
# Loop through each feature column
for column in df_noinstance:
    curr_feature = df_noinstance[column].tolist()
   
    # Get K samples of size n randomly, with replacement
    num_samples = 1000
    sample_size = 50
    sample_SD_list = []
 
    for x in range(0, num_samples):
        sample = np.random.choice(curr_feature, sample_size)
        sample_SD = statistics.stdev(sample)
        sample_SD_list.append(sample_SD)
        
    # Find lower and upper bound of 90% confidence interval
    lower_bound = np.quantile(sample_SD_list, 0.05)
    upper_bound = np.quantile(sample_SD_list, 0.95)
    
    all_lower_list.append(lower_bound)
    all_upper_list.append(upper_bound)    
    
# Create new dataframe for confidence interval
confidence_interval_df = pd.DataFrame({"Feature": SD_df_columns, "Standard Deviation": SD_list, "Confidence Interval Lower Bound": all_lower_list, "Confidence Interval Upper Bound": all_upper_list})
confidence_interval_df
```

#### I collect a set of data (n = 100 observations) containing a single predictor and a quantitative response. I then fit a linear regression model to the data, as well as a separate cubic regression, i.e. Y = 0 +1X +2X2 +3X3 +.

#### (a) Suppose that the true relationship between X and Y is linear, i.e. Y = 0 + 1X + . Consider the training residual sum of squares (RSS) for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

We would expect the training residual sum of squares (RSS) for the cubic regression to be lower than the training RSS for the linear regression, given that the true relationship is linear. Polynomial models like the cubic regression result in an estimate of f with more degrees of freedom and higher flexibility, which fit closer to the true function f and the noise, regardless of whether f is linear or nonlinear. In the training data, this would reduce the unexplained variation between the observed and predicted values of the dependent variable, which is measured by the RSS.

#### (b) Answer (a) using test rather than training RSS.

We would expect the test RSS for the linear regression to be lower than the test RSS for the cubic regression, given that the true relationship is linear. Because the cubic model is more flexible, it would pick up patterns in the train dataset caused by random chance rather than properties of the true unknown function f and overfit the data. The cubic model only performs well on the train data set, but will fail to fit the additional test dataset and the RSS will be higher than in a less flexible model.

#### (c) Suppose that the true relationship between X and Y is not linear, but we dont know how far it is from linear. Consider the training RSS for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

We would expect the training RSS for the cubic regression to be lower than the training RSS for the linear regression, given that the true relationship is not linear and we don't know how far it is from linear. Regardless of how nonlinear the true function f is, a polynomial model such as the cubic regression would result in a more flexible model with more degrees of freedom that would fit more accurately to a wider range of nonlinear shapes. In the training data, this would reduce the unexplained variation between the observed and predicted values of the dependent variable, which is measured by the RSS. 

#### (d) Answer (c) using test rather than training RSS.

There is not enough information to determine whether the test RSS for the cubic regression is higher or lower than the test RSS for the linear regression. Whether the test RSS for the cubic or linear regression is lower depends on how far away from linear the true relationship between the variables is, which is unknown. If the true function is closer to linear, the linear regression would prevent overfitting on the training data, perform better on the test data, and result in a lower RSS and lower unexplained variation between the observed and predicted values of the dependent variable. If the true function is more nonlinear, the cubic model would have more degrees of freedom and be flexible enough to capture the nonlinearities and result in a lower RSS.
